```python
import gmaps
import gmaps.datasets
from config import gmaps_api
import numpy as np
import pandas as pd
gmaps.configure(api_key=gmaps_api)
```

```python
breweries_csv = pd.read_csv('Code/brewery_data.csv')
breweries_df = pd.DataFrame(breweries_csv)
breweries_df
```

```python
breweries_df.columns
```

```python
# Need to get rid of NaN lat and lon in order to plot

breweries_val = breweries_df.dropna(subset=['latitude', 'longitude'])
breweries_val
```

```python
locations = breweries_val[['latitude', 'longitude']]
```

```python
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations))
fig

# Click the download arrow in the top left corner to download the map as a PNG
# gmaps does not currently support saving as PNG [https://jupyter-gmaps.readthedocs.io/en/latest/export.html]
```
